```python
# creating a dataframe using the random package 

# first generate data 
age = np.random.choice([15, 20, 30, 45, 12, '10', 15, '34', 7, 'Missing'], 50)
fico = np.random.choice(['100-150', '150-200', '200-250', '250-300'],50)
city = np.random.choice(['Mumbai', 'Delhi', 'Chennai', 'Kolkata'],50)
ID = np.arange(50)
rating = np.random.choice(['Excellent', 'Good', 'Bad', 'Pathetic'], 50)
balance = np.random.choice([10000, 20000, 30000, np.nan, 50000, 60000], 50)
children = np.random.randint(high=5,low=0,size=(50,))

```

```python
# stacking / combining data frames by rows - tis will duplocate the row indices
df1 = pd.DataFrame([['a', 1], ['b', 2]], columns= ['letter','number'])
df2 = pd.DataFrame([['a', 1, 'cat'], ['b', 2, 'dog']], columns= ['letter','number', 'animal'])
pd.concat([df1,df2], axis = 0)
```

```python
# this will help prevent duplication of indices
pd.concat([df1,df2], axis = 0, ignore_index=True)
```

```python
# to combine data frames using columns
df1 = pd.DataFrame([['tiger', 'leo'], ['lion', 'busa'], ['cheetah', 'rex']], columns= ['animal','name'])
df2 = pd.DataFrame([['a', 1, 'cat'], ['b', 2, 'dog']], columns= ['letter','number', 'animal'])
pd.concat([df1,df2], axis = 0)
```

```python
# merging data frames like sql joins

df1=pd.DataFrame({'cust_id':[1,2,3,4,5],'product':['radio','pen','pencil','eraser','mobile']})
df2=pd.DataFrame({'cust_id':[3,4,5,6,7],'state':['MH','MH','GJ','MP','UP']})

df1
df2
```

```python
# right join will give all the columns from the right table
pd.merge(df1,df2,on=['cust_id'],how='right')
```

```python
# left join will give all the columns from the left table
pd.merge(df1,df2,on=['cust_id'],how='left')
```

```python
# inner join will give all the commmon columns from both tables
pd.merge(df1,df2,on=['cust_id'],how='inner')
```
